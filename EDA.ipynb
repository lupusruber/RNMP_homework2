## Basic EDA for the dataset

```python
import pandas as pd
```

```python
main_data_df = pd.read_csv('ml-100k/u.data', sep=r'\t', engine='python')
main_data_df.timestamp = pd.to_datetime(main_data_df.timestamp, unit='s')
main_data_df = main_data_df.sort_values(by='timestamp', ascending=True).reset_index(drop=True)
```
